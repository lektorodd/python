## Holde Anaconda oppdatert

Dei ulike biblioteka og applikasjonane i Anaconda-pakken er under stadig utvikling, og det kan vera lurt å oppdatera pakken med jamne mellomrom. Dette er enklast å gjera i terminalen. Under er ein guide for korleis du gjer dette på Mac og på Windows. 

### Mac

For å opna terminalen kan du trykka på forstørrelsesglaset på statuslinja øverst (hurtigtast: command + space), skriva inn _Terminal_ og trykka enter. 

Deretter skriv du inn
```bash
conda update --all
```
og trykker enter. På eit tidspunkt får du spørsmål om du vil halda fram (y/n). Skriv inn _y_ (for yes) og trykk enter. 

::::{admonition} Tips
:class: tip

For å installere den _aller_ nyaste versjonen (pr. no v3.3.0) kan det hende at du må bruka 
```bash
conda install -c conda-forge jupyterlab
```
men du bør først ha oppdatert alt som skildra over. 

For å sjekka versjonen av Jupyter Lab kan du skriva
```bash
jupyter lab --version
``` 
i terminalen. 
::::

### Windows

==Fiks opp i dette==
